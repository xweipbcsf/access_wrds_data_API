```python
cen_inst_reco = pd.merge(cen_inst, reco_2015, how = 'inner', on = 'cusip_8')
```

```python
cen_inst_reco.head().T
```

```python
len(cen_inst_reco['group'])
```

```python
cen_inst_reco['in_diff'] = cen_inst_reco['in_degree_2016'] - cen_inst_reco['in_degree_2015']
```

```python
cen_inst_reco = pd.read_pickle('cen_inst_reco.pkl')
```

```python
df1 = cen_inst_reco.loc[cen_inst_reco['out_diff'] < 0]
df1['group'] = -1
```

```python
df2 = cen_inst_reco.loc[cen_inst_reco['out_diff'] == 0]
df2['group'] = 0
```

```python
df3 = cen_inst_reco.loc[cen_inst_reco['out_diff'] > 0]
df3['group'] = 1
```

```python
cen_inst_reco = pd.concat([df1,df2,df3])
```

```python
cen_inst_reco['out_diff_ratio'] = cen_inst_reco['out_diff']/cen_inst_reco['out_degree_2015']
```

```python
cen_inst_reco.info()
```

```python
cen_inst_reco['out_diff_ratio'] = cen_inst_reco['out_diff_ratio'].replace([np.inf, -np.inf], 0)
```

```python
cen_inst_reco = cen_inst_reco.fillna(0)
```

```python
cen_inst_reco.to_pickle('cen_inst_reco.pkl')
```
